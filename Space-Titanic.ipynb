```python
import pandas as pd
```

```python
df = pd.read_csv("spaceship-titanic\\train.csv") # (8693, 14)

df.head()
```

```python
df.Destination.value_counts()
```

```python
df.HomePlanet.value_counts()
```

```python
# to drop : PassengerId,Cabin,Name
```

```python
df.info()
```

```python
df1 = df.drop(columns=['PassengerId','Cabin','Name'],axis=1)
df1.head()
```

```python
# Dealing to Nan Values
df1.HomePlanet = df1.HomePlanet.fillna(df1.HomePlanet.mode()[0])
df1.CryoSleep = df1.CryoSleep.fillna(False)
df1.Destination = df1.Destination.fillna(df1.Destination.mode()[0])
df1.Age = df1.Age.fillna(df1.Age.median())
df1.VIP = df1.VIP.fillna(False)
df1.RoomService = df1.RoomService.fillna(df1.RoomService.median())
df1.FoodCourt = df1.FoodCourt.fillna(df1.FoodCourt.median())
df1.ShoppingMall = df1.ShoppingMall.fillna(df1.ShoppingMall.median())
df1.Spa = df1.Spa.fillna(df1.Spa.median())
df1.VRDeck = df1.VRDeck.fillna(df1.VRDeck.median())

df1.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in df1.columns:
    if df1[col].dtypes =='O' or df1[col].dtypes=='bool':
        df1[col] = le.fit_transform(df1[col])

df1.head()
```

```python
df1.Transported.value_counts()
df1["Total Fare"] = df1['RoomService'] + df1['FoodCourt'] + df1['ShoppingMall'] + df1['Spa'] + df1['VRDeck']
df1
```

```python
x = df1.drop(columns=['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck',"Transported"])
y = df1["Transported"]
x
```

```python
test = pd.read_csv("spaceship-titanic\\test.csv")
test.info()
```

```python
PassengerIdcol = test['PassengerId']
test  = test.drop(columns=['PassengerId','Cabin','Name'],axis=1)
test.HomePlanet = test.HomePlanet.fillna(test.HomePlanet.mode()[0])
test.CryoSleep = test.CryoSleep.fillna(False)
test.Destination = test.Destination.fillna(test.Destination.mode()[0])
test.Age = test.Age.fillna(test.Age.median())
test.VIP = test.VIP.fillna(False)
test.RoomService = test.RoomService.fillna(test.RoomService.median())
test.FoodCourt = test.FoodCourt.fillna(test.FoodCourt.median())
test.ShoppingMall = test.ShoppingMall.fillna(test.ShoppingMall.median())
test.Spa = test.Spa.fillna(test.Spa.median())
test.VRDeck = test.VRDeck.fillna(test.VRDeck.median())
for col in test.columns:
    if test[col].dtypes =='O' or test[col].dtypes=='bool':
        test[col] = le.fit_transform(test[col])

test["Total Fare"] = test['RoomService'] + test['FoodCourt'] + test['ShoppingMall'] + test['Spa'] + test['VRDeck']
x_test = test.drop(columns=['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck'])

x_test.head()
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 10000)
lr.fit(x,y)
```

```python
y_pred = lr.predict(x_test)
```

```python
df_submission = pd.DataFrame(y_pred.astype(bool))


df_submission["PassengerId"] = PassengerIdcol
df_submission.columns = ['Transported', 'PassengerId']

df_submission = df_submission.reindex(columns=["PassengerId","Transported"])
df_submission.to_csv("Submission_titanic_space.csv",index=False)
```
